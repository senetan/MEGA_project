```python
def extract_examples(dataset, dataset_name):
    print(f"\n=== Examples for {dataset_name} ===")

    # Print the 'covered_area' field (which is a dictionary)
    covered_area = dataset.get("covered_area", {})
    print("\nCovered_area example:")
    print(json.dumps(covered_area, indent=2))

    # Process the 'history' field if available (it's expected to be a list)
    history = dataset.get("history", [])
    if history:
        print("\nHistory examples:")
        datetimes = []
        # Extract download_datetime values and convert to datetime objects
        for record in history:
            dt_str = record.get("payload", {}).get("download_datetime")
            if dt_str:
                try:
                    # Parse the datetime string (example format: "2025-03-15T06:09:03.376345Z")
                    dt_obj = datetime.strptime(dt_str, "%Y-%m-%dT%H:%M:%S.%fZ")
                    datetimes.append(dt_obj)
                except Exception as e:
                    print(f"Error parsing date '{dt_str}': {e}")
        if datetimes:
            earliest = min(datetimes)
            latest = max(datetimes)
            print("Earliest download_datetime:", earliest.isoformat() + "Z")
            print("Latest download_datetime:", latest.isoformat() + "Z")
        else:
            print("No valid 'download_datetime' values found in history records.")
    else:
        print("\nNo history data available.")

# Assuming 'gas_stations' and 'ev_charging' are already defined dictionaries:
extract_examples(gas_stations, "Gas Stations Dataset")
extract_examples(ev_charging, "EV Charging Stations Dataset")
```
